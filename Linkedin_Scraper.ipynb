```python
from urllib.request import urlopen
from urllib.parse import urlparse
import urllib
from bs4 import BeautifulSoup
import requests
import pandas as pd
import sys
from time import sleep
import requests
from lxml import html
import sys
import os
from google.colab import drive
import traceback
```

```python
url='https://www.linkedin.com/jobs/search/?f_TPR=r86400&geoId=102713980&location=India&start='
```

```python
name=[]
company=[]
location=[]
time=[]
link=[]
```

```python
opener = urllib.request.build_opener()
opener.addheaders = [('User-agent', 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.102 Safari/537.36')]
urllib.request.install_opener(opener)
for x in range(0,2000,25):
  try:
    tempurlll=url+str(x)
    print(x)
    response = urllib.request.urlopen(tempurlll)
    soup = BeautifulSoup(response, 'html.parser')
    raw_urls = soup.find_all('ul', {'class':'jobs-search__results-list'})
    asd=list(raw_urls[0].find_all("li"))
    for i in asd:
      try:
        tempname=i.find('h3', {'class': 'base-search-card__title'}).get_text().strip()
        tempcomp=i.find('a', {'class': 'hidden-nested-link'}).get_text().strip()
        temploc=i.find('span', {'class': 'job-search-card__location'}).get_text().strip()
        temptime=i.find('time', {'class': 'job-search-card__listdate--new'}).get_text().strip()
        templink=i.find('a', {'class': 'base-card__full-link absolute top-0 right-0 bottom-0 left-0 p-0 z-[2]'}).get('href')

        name.append(tempname)
        company.append(tempcomp)
        location.append(temploc)
        time.append(temptime)
        link.append(templink)
        
      except:
        print(traceback.format_exc())
        pass
    sleep(2)
  except:
    print(traceback.format_exc())
    pass
```

```python
finaldf=pd.DataFrame.from_dict({'Name':name, 'Company': company, 'Location': location, 'Time': time, 'Link': link})
```

```python
finaldf.to_csv("finaldf.csv", index=False)
```
